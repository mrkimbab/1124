```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# chrome 키기
crom = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 긁어올 스팀 일일 최다 플레이 차트
url = "https://store.steampowered.com/charts/mostplayed"
crom.get(url)

time.sleep(5)

# 테이블에서 모든 행을 찾기
table_rows = crom.find_elements(By.CSS_SELECTOR, 'table > tbody > tr')

for row in table_rows:
    try:
        # 동시접속자 , 최다 동시 접속자, 가격 클래스 뽑아오기
        player_count = row.find_element(By.CSS_SELECTOR, 'td:nth-child(5)').text.strip()
        daily_highest_player_count = row.find_element(By.CSS_SELECTOR, 'td:nth-child(6)').text.strip()
        price = row.find_element(By.CSS_SELECTOR, 'td:nth-child(4)').text.strip()
        
        # 게임 이름
        game_name = row.find_element(By.CSS_SELECTOR, 'td > a').text.strip()
        # 게임 링크
        game_link = row.find_element(By.CSS_SELECTOR, 'td > a').get_attribute('href')

        # 게임 페이지로 이동, 태그 정보긁기
        crom.get(game_link)
        time.sleep(3)

        # 태그 긁기
        try:
            tags = crom.find_elements(By.CSS_SELECTOR, '.app_tag')
            tag_list = [tag.text for tag in tags]
        except Exception as tag_e:
            tag_list = []  # 태그를 찾지 못할 경우 빈 리스트

        # 출력
        print(f"게임: {game_name}, 동시 접속자 수: {player_count}, 최고 동시 접속자 수: {daily_highest_player_count}, 가격: {price}, 태그: {tag_list}")
        
        # 다시 돌아가기
        crom.back()
        time.sleep(3)

    except Exception as e:
        print()

crom.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv

crom = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

url = "https://store.steampowered.com/charts/mostplayed"
crom.get(url)

time.sleep(5)  # 페이지 로딩 시간 충분히 주기

# CSV 파일 작성
with open('steam_games_chart1.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Game Name', 'Player Count', 'Daily Highest Player Count', 'Price', 'Tags'])

    # 테이블에서 모든 게임 정보 추출
    table_rows = crom.find_elements(By.CSS_SELECTOR, 'table > tbody > tr')

    for row in table_rows:
        try:
            player_count = row.find_element(By.CSS_SELECTOR, 'td:nth-child(5)').text.strip()
            daily_highest_player_count = row.find_element(By.CSS_SELECTOR, 'td:nth-child(6)').text.strip()
            price = row.find_element(By.CSS_SELECTOR, 'td:nth-child(4)').text.strip()

            game_name = row.find_element(By.CSS_SELECTOR, 'td > a').text.strip()

            game_link = row.find_element(By.CSS_SELECTOR, 'td > a').get_attribute('href')

            # 게임 페이지로 이동
            crom.get(game_link)
            time.sleep(3)  # 게임 페이지 로딩 대기

            try:
                # 태그 로드 대기 (10초)
                WebDriverWait(crom, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.app_tag')))
                
                # 태그를 다시 찾기 (stale element reference 오류 방지)
                tags = crom.find_elements(By.CSS_SELECTOR, '.app_tag')
                tag_list = [tag.text for tag in tags][:8]  # 태그를 8개 추출
            except Exception as e:
                print()
                tag_list = []  # 오류가 발생하면 태그는 빈 리스트로 설정

            # 출력, CSV 저장
            print(f"게임: {game_name}, 동시 접속자 수: {player_count}, 최고 동시 접속자 수: {daily_highest_player_count}, 가격: {price}, 태그: {tag_list}")
            writer.writerow([game_name, player_count, daily_highest_player_count, price, ", ".join(tag_list)])

            # 다시 메인 페이지로 돌아가기
            crom.back()
            time.sleep(3)  # 페이지 전환 대기

        except Exception as e:
            print()

crom.quit()
```
