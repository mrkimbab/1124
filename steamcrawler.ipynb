```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# chrome 키기
crom = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 긁어올 스팀 일일 최다 플레이 차트
url = "https://store.steampowered.com/charts/mostplayed"
crom.get(url)

time.sleep(5)

# 테이블에서 모든 행을 찾기
table_rows = crom.find_elements(By.CSS_SELECTOR, 'table > tbody > tr')

for row in table_rows:
    try:
        # 동시접속자 , 최다 동시 접속자, 가격 클래스 뽑아오기
        player_count = row.find_element(By.CSS_SELECTOR, 'td:nth-child(5)').text.strip()
        daily_highest_player_count = row.find_element(By.CSS_SELECTOR, 'td:nth-child(6)').text.strip()
        price = row.find_element(By.CSS_SELECTOR, 'td:nth-child(4)').text.strip()
        
        # 게임 이름
        game_name = row.find_element(By.CSS_SELECTOR, 'td > a').text.strip()
        # 게임 링크
        game_link = row.find_element(By.CSS_SELECTOR, 'td > a').get_attribute('href')

        # 게임 페이지로 이동, 태그 정보긁기
        crom.get(game_link)
        time.sleep(3)

        # 태그 긁기
        try:
            tags = crom.find_elements(By.CSS_SELECTOR, '.app_tag')
            tag_list = [tag.text for tag in tags]
        except Exception as tag_e:
            tag_list = []  # 태그를 찾지 못할 경우 빈 리스트

        # 출력
        print(f"게임: {game_name}, 동시 접속자 수: {player_count}, 최고 동시 접속자 수: {daily_highest_player_count}, 가격: {price}, 태그: {tag_list}")
        
        # 다시 돌아가기
        crom.back()
        time.sleep(3)

    except Exception as e:
        print()

crom.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv

crom = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

url = "https://store.steampowered.com/charts/mostplayed"
crom.get(url)

time.sleep(5)  # 페이지 로딩 시간 충분히 주기

# CSV 파일 작성
with open('steam_games_chart1.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Game Name', 'Player Count', 'Daily Highest Player Count', 'Price', 'Tags'])

    # 테이블에서 모든 게임 정보 추출
    table_rows = crom.find_elements(By.CSS_SELECTOR, 'table > tbody > tr')

    for row in table_rows:
        try:
            player_count = row.find_element(By.CSS_SELECTOR, 'td:nth-child(5)').text.strip()
            daily_highest_player_count = row.find_element(By.CSS_SELECTOR, 'td:nth-child(6)').text.strip()
            price = row.find_element(By.CSS_SELECTOR, 'td:nth-child(4)').text.strip()

            game_name = row.find_element(By.CSS_SELECTOR, 'td > a').text.strip()

            game_link = row.find_element(By.CSS_SELECTOR, 'td > a').get_attribute('href')

            # 게임 페이지로 이동
            crom.get(game_link)
            time.sleep(3)  # 게임 페이지 로딩 대기

            try:
                # 태그 로드 대기 (10초)
                WebDriverWait(crom, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.app_tag')))
                
                # 태그를 다시 찾기 (stale element reference 오류 방지)
                tags = crom.find_elements(By.CSS_SELECTOR, '.app_tag')
                tag_list = [tag.text for tag in tags][:8]  # 태그를 8개 추출
            except Exception as e:
                print()
                tag_list = []  # 오류가 발생하면 태그는 빈 리스트로 설정

            # 출력, CSV 저장
            print(f"게임: {game_name}, 동시 접속자 수: {player_count}, 최고 동시 접속자 수: {daily_highest_player_count}, 가격: {price}, 태그: {tag_list}")
            writer.writerow([game_name, player_count, daily_highest_player_count, price, ", ".join(tag_list)])

            # 다시 메인 페이지로 돌아가기
            crom.back()
            time.sleep(3)  # 페이지 전환 대기

        except Exception as e:
            print()

crom.quit()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

df = pd.read_csv('steam_games_chart1.csv', encoding='cp1252')

# 태그 데이터 추출
tags = df['Tags'].dropna()

# 태그 조건 처리
def process_tags(tags):
    # 쉼표(,)로 구분된 태그를 하나의 문자열로 결합하고, 각 태그를 공백을 _로 변환
    processed_tags = []
    for tag in tags.split(','):
        tag = tag.strip()  # 태그 앞뒤 공백 제거
        if tag.startswith('_'):
            # _로 시작하는 태그는 '_'를 제거한 후 동일한 태그로 처리
            tag = tag[1:]  # 첫 번째 _를 제거
        tag = tag.replace(" ", "_")  # 공백을 _로 바꾸기
        processed_tags.append(tag)
    return processed_tags

# 각 태그에 대해 조건 처리 후 유효옵만 추출
all_tags = []
for tag_list in tags:
    processed_tag_list = process_tags(tag_list)
    all_tags.extend(processed_tag_list)

tags_text = ' '.join(all_tags)

# 워드 클라우드 객체 생성
wordcloud = WordCloud(width=800, height=400, background_color="white", collocations=True).generate(tags_text)

# 워드 클라우드 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  # 축 숨기기
plt.show()
```

```python
# 해당 코드는 공백 태그를 포함하고 있어 오류가 남, 다음 코드를 사용

import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from itertools import combinations

df = pd.read_csv('steam_games_chart1.csv', encoding='cp1252')

tags = df['Tags'].dropna()

def process_tags(tags):
    processed_tags = []
    for tag in tags.split(','):
        tag = tag.strip()
        if tag.startswith('_'):
            tag = tag[1:] 
        tag = tag.replace(" ", "_") 
        processed_tags.append(tag)
    return processed_tags

tag_combinations = []
for tag_list in tags:
    processed_tag_list = process_tags(tag_list)
    tag_combinations.extend(combinations(processed_tag_list, 2))  # 태그 쌍 생성

# 태그 빈도
tag_counter = Counter(tag_combinations)

# 결과를 데이터프레임으로 변환
tp_df = pd.DataFrame(tag_counter.items(), columns=['Tag Pair', 'Count'])
tp_df = tp_df.sort_values(by='Count', ascending=False)

# 가장많이 나온 태그 20개
highestpairs = tp_df.head(20)

# x, y 좌표 크기
x_coords = highestpairs['Tag Pair'].apply(lambda x: x[0])  # 첫 번째 태그 x 좌표
y_coords = highestpairs['Tag Pair'].apply(lambda x: x[1])  # 두 번째 태그 y 좌표
sizes = highestpairs['Count'] * 30

# 산점도
plt.figure(figsize=(10, 6))
plt.scatter(x_coords, y_coords, s=sizes, color='blue', alpha=0.5, edgecolors='w', marker='o')

plt.title('태그 상호 연관성')

# 그래프 표시
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from itertools import combinations

df = pd.read_csv('steam_games_chart1.csv', encoding='cp1252')

tags = df['Tags'].dropna()

# 태그 조건 처리
def process_tags(tags):
    processed_tags = []
    for tag in tags.split(','):
        tag = tag.strip()
        if not tag: # 태그가 공백태그일경우 스킵
            continue
        if tag.startswith('_'):
            tag = tag[1:] 
        tag = tag.replace(" ", "_") 
        processed_tags.append(tag)
    return processed_tags

tag_combinations = []
for tag_list in tags:
    processed_tag_list = process_tags(tag_list)
    tag_combinations.extend(combinations(processed_tag_list, 2))  # 태그 쌍 생성

# 태그 빈도
tag_counter = Counter(tag_combinations)

# 결과를 데이터프레임으로 변환
tp_df = pd.DataFrame(tag_counter.items(), columns=['Tag Pair', 'Count'])
tp_df = tp_df.sort_values(by='Count', ascending=False)

# 가장많이 나온 태그 20개
highestpairs = tp_df.head(20)

x_coords = highestpairs['Tag Pair'].apply(lambda x: x[0])  # 첫 번째 태그 x 좌표
y_coords = highestpairs['Tag Pair'].apply(lambda x: x[1])  # 두 번째 태그 y 좌표
sizes = highestpairs['Count'] * 50

# 산점도 그리기 (크기와 색상 설정)
plt.figure(figsize=(10, 6))
plt.scatter(x_coords, y_coords, s=sizes, color='blue', alpha=0.5, edgecolors='w', marker='o')

plt.title('태그 상호 연관성')

# 그래프 표시
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
